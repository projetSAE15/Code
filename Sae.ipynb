```python
import pandas
import numpy
```

```python
s = pandas.Series([1, 2, 5, 7])
print(s)
```

```python
s = pandas.Series([1, 2, 5, 7], dtype = numpy.int8)
print(s)
```

```python
s = pandas.Series(['a', 'b', 'a', 'a', 'c'], dtype = 'category')
print(s)
```

```python
s = pandas.Categorical(['a', 'b', 'a', 'a', 'c'], categories = ['c', 'b', 'a'], ordered = True)
print(s)
```

```python
s = pandas.Series([1, 5, 3, 9, numpy.inf, numpy.nan])
s[s < 5] = 99 
s[s == numpy.inf] = 99
s[numpy.isnan(s)] = 99 
print(s)
```

```python
s = pandas.Series(['a', 'b', 'a', 'a', 'c'], dtype = 'category')
s.replace({'a': 'A', 'b': 'B'})
```

```python
s = pandas.Series(['a', 'b', 'a', 'a', 'c'], dtype = 'category')
s.map({'a': 'A', 'b': 'B'})
```

```python
s=pandas.Series([1, 2, 5, 7], index = ['a', 'b', 'c', 'd'])
print(s)
```

```python
s=pandas.Series({'a': 1, 'b': 2, 'c': 5, 'd': 7})
print(s)
#c'est avec un dico / revient au même que l'exemple juste au-dessus 
#(si on donne un dictionnaire, les éléments sont triés par ordre alphabétique des clefs).
```
